```python
song_scores = [72, 91, 88, 64, 95, 81, 77, 89, 93, 68]
song_titles = [
    "Midnight Drive",
    "Golden Hour",
    "Neon Lights",
    "Falling Slowly",
    "Electric Heart",
    "Ocean Eyes",
    "Firestorm",
    "Starlight",
    "Runaway",
    "Dreamscape"
]

hit = 85
hit_scores = []
hit_titles = []

for x in range(len(song_scores)):
    if song_scores[x] >= hit:
        hit_scores.append(song_scores[x])
        hit_titles.append(song_titles[x])

print(hit_scores)
print(hit_titles)
```

Question 2:
Using your hit_scores list from Part (a):

Convert it into a NumPy array called hit_scores_arr.

Compute:

num_hits → total number of hits

avg_hits → average popularity

min_hits → lowest popularity

Create:

centered_hits → subtract the mean from each score (one operation).

Create a Boolean array:

below_average → scores strictly less than the mean.

```python
hit_scores_array = numpy.array(hit_scores)

num_hits = len(hit_scores_array)
print("Number of Hits: ", num_hits)

sum = 0
for value in hit_scores:
    sum += value
avg_hits = sum / num_hits
print("Average Score of Hits: ", avg_hits)

min_hits = 100
for value in hit_scores:
    if value < min_hits:
        min_hits = value
print("Minimum Hit Score: ", min_hits)

centered_hits = []
x = 0
for value in hit_scores:
    x = value - avg_hits
    centered_hits.append(x)
print(centered_hits)

below_average = []
boolean = False
for value in hit_scores:
    if value < avg_hits:
        boolean = True
    below_average.append(boolean)
below_average = numpy.array(below_average)
print(below_average)
```
