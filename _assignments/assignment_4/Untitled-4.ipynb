Question 1:
Write Python code that assigns a letter grade to the variable grade based on the variable score using this scale:

93 and above → "A"

87–92 → "A-"

83–86 → "B+"

Anything below 83 → "Below B+"

Use an if / elif / else structure.

```python
score = numpy.random.randint(high = 101, low = 0)
print(score)

if score >= 93:
    print("A")
elif 87 <= score <= 92:
    print("A-")
elif 83 <= score <= 86:
    print("B+")
else:
    print("Below B+")
```

Question 2:
Creates the list

numbers = [12, 7, 25, 30, 4, 19]

Uses a for loop to build a new list called even_numbers.

Only adds numbers that are even.

Prints even_numbers at the end.

```python
numbers = [12, 7, 25, 30, 4, 19]
even_numbers = []

for value in numbers:
    if value % 2 == 0:
        even_numbers.append(value)

print(even_numbers)
```

Question 4:
Write code that:

Uses these two lists:

names = ["Ava", "Ben", "Chloe", "Daniel"]
scores = [91, 76, 88, 64]

Prints the name of each student who scored 85 or higher.

```python
names = ["Ava", "Ben", "Chloe", "Daniel"]
scores = [91, 76, 88, 64]

for names, scores in zip(names, scores):
        if scores >= 85:
            print(names)
```

Question 7:
Write code that:

```python
number = 10

while number != 0:
    print(number)
    number = number - 1
```

Question 13:
Uses this list:

values = [4, 9, 12, 3, 18]

Uses a while loop (not a for loop).

Prints only the values that are divisible by 3.

Uses indexing (values[i]) inside the loop

```python
values = [4, 9, 12, 3, 18]

count = len(values)
x = 0

while x < count:
    for value in values:
        if value % 3 == 0:
            print(value)
```
